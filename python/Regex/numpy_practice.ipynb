Multi dimentional Array

```python
dimentional=np.array((['jan','feb','mar'],
                    [300,400,500],
                    [600,700,800],
                    [50,30,200]))
print(dimentional.shape)
print(dimentional)
```

```python
mc=np.array((['jan','feb','mar'],
                    [300,400,500],
                    [600,700,800],
                    [50,30,200]))
print(mc[1,:])
print(mc[:,1])
print(mc[1:,2])
print(mc[1:,1])
print(mc[1:,0])
print(mc[3:,])
```

```python
#satisfaction rating
sr=np.array([3.6,4,2,3,7,9])
print(sr[sr>3])
print(sr)
print(sr[(sr>=2) & (sr<=4)])
```

```python
mc=np.array((['jan','feb','mar'],
                    [300,400,500],
                    [600,700,800],
                    [50,30,200]))
print(np.sum(mc[1:].astype(int)))
```

```python
mc=np.array((['jan','feb','mar'],
                    [300,400,500],
                    [600,700,800],
                    [50,30,200]))
print(np.mean(mc[1:].astype(int)))
print(np.min(mc[1:].astype(int)))
print(np.max(mc[1:].astype(int)))
```

```python
north=np.array([5,100,30])
south=np.array([3,2,5])
print(np.concatenate((north,south)))
```

```python
mc=np.array((['jan','feb','mar'],
                    [300,400,500],
                    [600,700,800],
                    [50,30,200]))
print(mc[1:,])
calls=mc[1:,].astype(int)
print(np.add.reduce(calls,axis=0))
```

```python
import numpy as np

mc = np.array([
    ['jan', 'feb', 'mar'],
    [300, 400, 500],
    [600, 700, 800],
    [50, 30, 200]
])

result = np.sum(mc[1:].astype(int), axis=0)
print(result)
```
