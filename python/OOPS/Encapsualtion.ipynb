```python
class Employee:
    def __init__(self):
        self.__salary = 50000

    def get_salary(self):# get methoi
        return self.__salary
    
    def update_salary(self,salary):
        if salary>0:
         self.__salary = salary
         return "salary Updated"
        else:
         return "invalid salary"
    
Emp = Employee()
print(Emp.get_salary())
print(Emp.update_salary(10000))
print(Emp.get_salary())
```
